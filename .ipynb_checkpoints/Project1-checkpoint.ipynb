```python
import pandas as pd
import numpy as np

# Load data
data = pd.read_csv('sample.csv')
data['pickup_datetime'] = pd.to_datetime(data['pickup_datetime'])
data['dropoff_datetime'] = pd.to_datetime(data['dropoff_datetime'])

# Sort data by taxi/driver and pickup datetime
data.sort_values(by=['hack_license', 'pickup_datetime'], inplace=True)

# Compute idle time
data['next_pickup'] = data.groupby('hack_license')['pickup_datetime'].shift(-1)
data['idle_time'] = (data['next_pickup'] - data['dropoff_datetime']).dt.total_seconds()

# Filter out idle times greater than 4 hours (14400 seconds)
data = data[data['idle_time'] <= 14400]

# Calculate trip time and total occupied time by taxi/driver
data['trip_time'] = (data['dropoff_datetime'] - data['pickup_datetime']).dt.total_seconds()
total_occupied_time = data.groupby('hack_license')['trip_time'].sum()

# Calculate total time from first pickup to last dropoff for each taxi/driver
first_pickup = data.groupby('hack_license')['pickup_datetime'].first()
last_dropoff = data.groupby('hack_license')['dropoff_datetime'].last()
total_time = (last_dropoff - first_pickup).dt.total_seconds()
```

```python
# Calculate utilization
utilization = total_occupied_time / total_time
utilization = utilization.reset_index().rename(columns={0: 'utilization'})

utilization.head()
```

```python
import matplotlib.pyplot as plt

# Create a histogram of the utilization rates
plt.figure(figsize=(10, 6))
plt.hist(utilization['utilization'], bins=30, color='blue', alpha=0.7)
plt.title('Distribution of Utilization Rates')
plt.xlabel('Utilization Rate')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Create a bar chart for the top 5 and bottom 5 drivers by utilization
top_5 = utilization.nlargest(5, 'utilization')
bottom_5 = utilization.nsmallest(5, 'utilization')

plt.figure(figsize=(14, 7))
plt.bar(top_5['hack_license'], top_5['utilization'], color='green', label='Top 5 Drivers')
plt.bar(bottom_5['hack_license'], bottom_5['utilization'], color='red', label='Bottom 5 Drivers')
plt.title('Top 5 and Bottom 5 Drivers by Utilization Rate')
plt.xlabel('Driver License Hash')
plt.ylabel('Utilization Rate')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
import geopandas as gpd
from shapely.geometry import Point, shape
import json

# Load your taxi ride data
data = pd.read_csv('sample.csv')
data['pickup_datetime'] = pd.to_datetime(data['pickup_datetime'])
data['dropoff_datetime'] = pd.to_datetime(data['dropoff_datetime'])

# Load and prepare the GeoJSON data
with open('nyc-boroughs.geojson', 'r') as file:
    geojson_data = json.load(file)

# Convert GeoJSON to a GeoDataFrame
boroughs_gdf = gpd.GeoDataFrame.from_features(geojson_data['features'])

# Function to find the borough based on coordinates
def find_borough(lon, lat, geo_df):
    point = Point(lon, lat)
    for _, row in geo_df.iterrows():
        if row['geometry'].contains(point):
            return row['borough']
    return None

# Enrich the taxi data with drop-off boroughs
data['dropoff_borough'] = data.apply(lambda row: find_borough(row['dropoff_longitude'], row['dropoff_latitude'], boroughs_gdf), axis=1)

# Calculate the time to next fare
data.sort_values(by=['hack_license', 'pickup_datetime'], inplace=True)
data['next_pickup'] = data.groupby('hack_license')['pickup_datetime'].shift(-1)
data['time_to_next_fare'] = (data['next_pickup'] - data['dropoff_datetime']).dt.total_seconds()

# Filter and group by borough to calculate the average time to next fare
valid_data = data[(data['time_to_next_fare'] >= 0) & (data['time_to_next_fare'] <= 14400)]  # Adjust filters as necessary
average_time_by_borough = valid_data.groupby('dropoff_borough')['time_to_next_fare'].mean()

# Output the results
print(average_time_by_borough)
```

```python
average_time_by_borough.plot(kind='bar', color='skyblue', figsize=(10, 6))

plt.title('Average Time to Next Fare by Drop-off Borough')
plt.xlabel('Borough')
plt.ylabel('Average Time to Next Fare (seconds)')
plt.xticks(rotation=45)
plt.grid(True)

plt.show()

```

```python
# Assuming data is already loaded
data = pd.read_csv('sample.csv')
data['pickup_datetime'] = pd.to_datetime(data['pickup_datetime'])
data['dropoff_datetime'] = pd.to_datetime(data['dropoff_datetime'])

# Load and prepare the GeoJSON data
with open('nyc-boroughs.geojson', 'r') as file:
    geojson_data = json.load(file)
boroughs_gdf = gpd.GeoDataFrame.from_features(geojson_data['features'])

# Define the find_borough function if not already defined
def find_borough(lon, lat, geo_df):
    point = Point(lon, lat)
    for _, row in geo_df.iterrows():
        if row['geometry'].contains(point):
            return row['borough']
    return None

# Enrich data with both pickup and dropoff boroughs
data['pickup_borough'] = data.apply(lambda row: find_borough(row['pickup_longitude'], row['pickup_latitude'], boroughs_gdf), axis=1)
data['dropoff_borough'] = data.apply(lambda row: find_borough(row['dropoff_longitude'], row['dropoff_latitude'], boroughs_gdf), axis=1)

# Ensure both columns are correctly populated
assert 'pickup_borough' in data.columns and 'dropoff_borough' in data.columns, "Borough columns not added successfully."

# Count trips that start and end in the same borough
data['same_borough'] = data['pickup_borough'] == data['dropoff_borough']
same_borough_count = data['same_borough'].sum()

# Output the result
print(f"Number of trips that start and end in the same borough: {same_borough_count}")
```

```python
data['different_borough'] = data['pickup_borough'] != data['dropoff_borough']
different_borough_count = data['different_borough'].sum()
print(f"Number of trips that start with one borough and end in different  borough: {different_borough_count}")
```
