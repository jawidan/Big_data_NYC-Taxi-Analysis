## Question A.1 (5 pts)

Let's start with dataset cleaning and data type conversions. 

- The `starttime` and `stoptime` are string types. Convert them into `datetime` types and round the values to seconds (e.g. `2020-03-01 08:00:30`).
- Convert `gender`, `usertype`, `start station name`, `end station name` to `category` dtype.
- Convert `start station id`, `end station id`, `bikeid` to `int32` dtype.
- Convert `birth year` to `int16` dtype.

````python
# 1. Convert starttime and stoptime to datetime types and round the values to seconds
df['starttime'] = pd.to_datetime(df['starttime']).dt.round('S')
df['stoptime'] = pd.to_datetime(df['stoptime']).dt.round('S')

# 2. Convert gender, usertype, start station name, end station name to category dtype.
df['gender'] = df['gender'].astype('category')
df['usertype'] = df['usertype'].astype('category')
df['start station name'] = df['start station name'].astype('category')
df['end station name'] = df['end station name'].astype('category')

# 3. Convert start station id, end station id, bikeid to int32 dtype.
df['start station id'] = df['start station id'].astype('int32')
df['end station id'] = df['end station id'].astype('int32')
df['bikeid'] = df['bikeid'].astype('int32')

# 4. Convert birth year to int16 dtype.
df['birth year'] = df['birth year'].astype('int16')

df.dtypes  
````

## Question A.2 (5 pts)

Plot trip duration distribution in a histogram, where each bin denotes 1 minute upto 1 hour (60 bins), and find out the most common trip duration in minutes.

````python
# 1. Convert trip duration from seconds to minutes
df['tripduration_minutes'] = df['tripduration'] / 60

# 2. Define the histogram
plt.figure(figsize=(10,6))
plt.hist(df['tripduration_minutes'], bins=60, range=[0,60])
plt.title('Histogram Trip Duration Distribution')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Most common trip duration in minutes
most_common_trip_duration = df['tripduration_minutes'].mode()[0]
print("Most common duration = " , most_common_trip_duration)
````

## Question A.3 (5 pts)

Identify the peak hours from the dataset and the most popular weekdays (week days with the most trips).

Support your conclusions by data visualization.

````python
df['hour'] = df['starttime'].dt.hour
df['weekday'] = df['starttime'].dt.day_name()

counts_by_hour = df['hour'].value_counts().sort_index()

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 10))

axes[0].bar(counts_by_hour.index, counts_by_hour.values)
axes[0].set_xticks(range(0, 24))
axes[1].set_title('Day trips')
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

counts_by_week = df['weekday'].value_counts().sort_index()

axes[1].bar(counts_by_week.index, counts_by_week.values)
axes[1].set_title('Week trips')
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

print("Peak hour is ",counts_by_hour.idxmax() ," and peak week is ", counts_by_week.idxmax())
````

## Question A.4 (5 pts)

Plot the age distribution (as of year 2020) of the bike users in a histogram, where each bin denotes one year. What are the most common age for men and women users, respectively? Are male users more than females?

**Requirements:**

- Exclude data from the users who are more than 80 years old. 
- Exclude data from the users who don't have a subscription (Customers with day passes).
- Exclude data from the users who don't have a binary gender (unknown gender).
- Plot distribution for men and women in a single figure with different colors.

````python
df['age_2020'] = 2020 - df['birth year']

filtered_df = df[
    (df['age_2020'] <= 80) &  
    (df['usertype'] == 'Subscriber') &  
    (df['gender'].isin([1, 2]))  
]

men = filtered_df[filtered_df['gender'] == 1]
women = filtered_df[filtered_df['gender'] == 2]


plt.figure(figsize=(12, 6))
plt.hist(men['age_2020'], bins=range(10, 81), color='green', alpha=0.6, label='Men')
plt.hist(women['age_2020'], bins=range(10, 81), color='yellow', alpha=0.6, label='Women')

plt.title('Graph')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

most_common_age_men = men['age_2020'].mode()[0]
most_common_age_women = women['age_2020'].mode()[0]

print("Most common men age is ",most_common_age_men , " while the value for woman is ", most_common_age_women)
````

## Question A.5 (6 pts)

Convert all station locations into shapely geometries, and visualize the stations (points) and its boroughs/counties (polygons). 

````python
stations = df.drop_duplicates(subset=['start station id', 'start station name', 'start station latitude', 'start station longitude'])
geometry = [Point(xy) for xy in zip(stations['start station longitude'], stations['start station latitude'])]
geo_stations = gpd.GeoDataFrame(stations, crs="EPSG:4326", geometry=geometry)

boroughs_path = 'dataset/borough.geojson'
ny_boroughs = gpd.read_file(boroughs_path)

geo_stations = geo_stations.to_crs(ny_boroughs.crs)

# Plotting the data
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
ny_boroughs.plot(ax=ax, color='lightblue', edgecolor='black')
geo_stations.plot(ax=ax, markersize=10, color='red', marker='o', label='Stations')
plt.title('NYC Bike Stations and Boroughs')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()

# In this task ChatGPT was used. 
````

## Question A.6 (6 pts)

Find those trips that have the same starting and ending stations (round trips), rank them by number of occurrences. Then, select top 20 trips, plot the stations on the map, and plot the number of occurrences in a bar chart. Why these stations have more round trips?

````python
round_trips = df[df['start station id'] == df['end station id']]

round_trip_counts = round_trips['start station name'].value_counts().head(20)

top_round_trip_stations = geo_stations[geo_stations['start station name'].isin(round_trip_counts.index)]

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 18))

axes[0].set_title('Top 20 Stations for Round Trips')
ny_boroughs.plot(ax=axes[0], color='lightblue', edgecolor='black')
top_round_trip_stations.plot(ax=axes[0], markersize=100 * (round_trip_counts / round_trip_counts.max()), 
                             color='red', marker='o', label='Stations')
axes[0].legend()

# Bar chart of the number of occurrences
axes[1].set_title('Number of Round Trips for Top 20 Stations')
round_trip_counts.plot(kind='bar', ax=axes[1], color='skyblue')
axes[1].set_xlabel('Station Name')
axes[1].set_ylabel('Number of Round Trips')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

round_trip_counts
````
